###  ***QUESTION 1: `How have GDP growth rates of the top 5 economies evolved over time?`***

```python
# Filter dataset for top 5
top5 = (df.groupby("country_name")["gdp_usd_cap"]
          .mean().sort_values(ascending=False).head(5).index)

d = df[df["country_name"].isin(top5)][["country_name", "year", "gdp_growth_cap"]].dropna()
```

```python
# Small Multiples (faceted line charts)
n = len(top5)
rows = int(np.ceil(n / 2))
cols = 2 if n > 1 else 1

fig = plt.figure(figsize=(12, 3.5 * rows))
for i, c in enumerate(top5, start=1):
    ax = fig.add_subplot(rows, cols, i)
    sub = d[d["country_name"] == c]
    ax.plot(sub["year"], sub["gdp_growth_cap"], marker="o", color="hotpink")
    ax.axhline(0, linewidth=1)
    ax.set_title(c)
    ax.set_xlabel("Year")
    ax.set_ylabel("GDP growth (%)")
    ax.grid(True, linewidth=0.3)
fig.suptitle("Trends in GDP Growth Rates among the Top 5 Economies (20102025)")
fig.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
# Bump Chart (rank by growth rate)
ranked = d.assign(rank=d.groupby("year")["gdp_growth_cap"].rank(method="first", ascending=False))

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(1,1,1)

for c in top5:
    sub = ranked[ranked["country_name"] == c].sort_values("year")
    ax.plot(sub["year"], sub["rank"], marker="o", linewidth=2)
    if not sub.empty:
        ax.text(sub["year"].max()+0.2, sub.iloc[-1]["rank"], c, ha="left", va="center")

ax.set_ylim(0.5, len(top5)+0.5)
ax.invert_yaxis()  # rank 1 at top
ax.set_yticks(range(1, len(top5)+1))
ax.set_xlabel("Year")
ax.set_ylabel("Rank of GDP Growth")
ax.set_title("Ranking of GDP Growth Rates among the Top 5 Economies (20102025)")
ax.grid(True, linewidth=0.3)
plt.tight_layout()
plt.show()
```

###  ***QUESTION 2: `What is the distribution of the unemployment rate per year across countries?`***

```python
df["year"] = pd.to_numeric(df["year"], errors="coerce")
df = df.dropna(subset=["year"])

# --- Focus years 20102024 ---
df_range = df[(df["year"] >= 2010) & (df["year"] <= 2024)].copy()

# --- Pick Top 10 by TOTAL GDP contribution over 20102024 ---
total_gdp = (
    df_range.groupby("country_name")["gdp_usd_cap"]
    .sum()
    .sort_values(ascending=False)
)
top10_countries = total_gdp.head(10).index.tolist()

pivot_abs = (
    df_range[df_range["country_name"].isin(top10_countries)]
    .pivot_table(index="year", columns="country_name", values="gdp_usd_cap", aggfunc="sum")
    .sort_index()
    .fillna(0.0)
)

col_totals = pivot_abs.sum(axis=0).sort_values(ascending=False)
pivot_abs = pivot_abs[col_totals.index.tolist()]

yearly_total_top10 = pivot_abs.sum(axis=1).replace(0, pd.NA)
pivot_share = (pivot_abs.div(yearly_total_top10, axis=0) * 100).dropna(how="all", axis=1)

legend_labels = [f"{c} ({col_totals[c]/1e12:.2f} T USD)" for c in pivot_abs.columns]
```

```python
# Line Chart
fig, ax = plt.subplots(figsize=(12, 7))
handles = []

for c in pivot_share.columns:
    y = pivot_share[c]
    if y.notna().sum() == 0:
        continue
    (h,) = ax.plot(pivot_share.index, y, marker="o", linewidth=2, label=c)
    handles.append(h)

ax.set_title("Share of Top-10 Total GDP by Country (20102024)  Line Chart", fontsize=14)
ax.set_xlabel("Year")
ax.set_ylabel("Share of Top-10 Total (%)")
ax.grid(True, linestyle="--", alpha=0.5)

#  Y-axis zoom: 3% to 20% for clearer differences
ax.set_ylim(3, 20)
ax.set_yticks(range(4, 21, 2))

#  Legend on the right with totals (sorted by total contribution)
ax.legend(
    handles,
    legend_labels[:len(handles)],  # guard in case some series were entirely NaN
    title="Country (Total GDP 20102024)",
    bbox_to_anchor=(1.25, 1),
    loc="upper left",
    borderaxespad=0.0,
)

# Reserve space on the right so legend is fully visible
plt.subplots_adjust(right=0.73)
plt.show()
```

```python
df["year"] = pd.to_numeric(df["year"], errors="coerce")
df = df.dropna(subset=["year"])

# === Identify Top 10 by TOTAL GDP contribution (sum across 20102024) ===
total_gdp = (
    df_range.groupby("country_name")["gdp_usd_cap"]
    .sum()
    .sort_values(ascending=False)
)

top10 = total_gdp.head(10)

df_top10 = df_range[df_range["country_name"].isin(top10.index)]
stack_pivot = (
    df_top10.pivot_table(
        index="year", columns="country_name", values="gdp_usd_cap", aggfunc="sum"
    )
    .sort_index()
    .fillna(0.0)
)
```

```python
# Stacked Area Chart

fig, ax = plt.subplots(figsize=(12, 7))
countries = stack_pivot.columns.tolist()
y_arrays = [stack_pivot[c] for c in countries]

poly_handles = ax.stackplot(stack_pivot.index, y_arrays, alpha=0.85)

ax.set_title("Stacked Area: Contribution to Total GDP (USD)  Top 10 Countries", fontsize=14)
ax.set_xlabel("Year")
ax.set_ylabel("GDP (USD)")
ax.grid(True, linestyle="--", alpha=0.4)

labels = [f"{c} ({total_gdp[c]/1e12:.2f} T USD)" for c in countries]

legend = ax.legend(poly_handles, labels,
                   title="Country (Total GDP 20102024)",
                   bbox_to_anchor=(1.25, 1), loc="upper left",
                   borderaxespad=0.)

plt.subplots_adjust(right=0.7)

plt.show()
```
